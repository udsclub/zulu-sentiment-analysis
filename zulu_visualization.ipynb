```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = DataFrame([['100k', 0.9082, 0.9492],
                 ['200k', 0.91562, 0.9529],
                 ['296k 50/50', 0.9416, 0.9656],
                 ['378k 50/50 + RT', 0.9186, 0.9510],
                 ['500k', 0.9218, 0.9562],
                 ['1M', 0.9263, 0.9587],
                 ['1,3M', 0.9298, 0.9606]], columns = ['n_rows', 'acc', 'f1'])
b = DataFrame([['100k', 0.6269, 0.7209],
                 ['200k', 0.6470, 0.7307],
                 ['296k 50/50', 0.7736, 0.7734],
                 ['378k 50/50 + RT', 0.8318, 0.8416],
                 ['500k', 0.6735, 0.746],
                 ['1M', 0.6951, 0.7589],
                 ['1,3M', 0.7022, 0.7623]], columns = ['n_rows', 'acc', 'f1'])
```

```python
plt.figure(figsize=((16,9)))
xs = [i for i, _ in enumerate(a.acc)]
plt.plot(xs, a.acc, 'g-', label='Accuracy')
plt.plot(xs, a.f1, 'r-', label='F1 score')
# plt.axis([0, 4, 0, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, a.n_rows)
plt.title("test on amazon data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
xs = [i for i, _ in enumerate(b.acc)]
plt.plot(xs, b.acc, 'g-', label='Accuracy')
plt.plot(xs, b.f1, 'r-', label='F1 score')
# plt.axis([0, 4, .9, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, b.n_rows)
plt.title("test on rt data")
plt.legend()
plt.show()
```
