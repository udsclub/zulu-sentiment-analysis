```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = DataFrame([['1:1', 0.9169, 0.9162],
                 ['1:1.5', 0.9184, 0.8916],
                 ['1:2.5', 0.9127, 0.8308],
                 ['1:5', 0.9231, 0.7119]], 
               columns = ['ratio', 'acc', 'f1'])
```

```python
plt.figure(figsize=((16,9)))
xs = [i for i, _ in enumerate(a.acc)]
plt.plot(xs, a.acc, 'g-', label='Accuracy')
plt.plot(xs, a.f1, 'r-', label='F1 score')
# plt.axis([0, 4, 0, 1.])
plt.xlabel("Ratio of negative samples to positive")
plt.xticks(xs, a.ratio)
plt.title("Test on amazon data")
plt.legend()
plt.show()
```
